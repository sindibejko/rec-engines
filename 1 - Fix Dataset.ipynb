```python
import pandas as pd
import numpy as np
import random

num_observations = 199999
base_weights = [1, 1.8, 3.4, 2.6, 1.2]
ratings_options = [1, 2, 3, 4, 5]

# Generate the ratings
ratings = [random.choices(ratings_options, weights=base_weights, k=1)[0] for _ in range(num_observations)]
```

```python
df['ratings'] = ratings
df.head()
```

```python
df = df[df['Action'] != 'purchase']
```

```python
df = df.drop(columns=['ZeroColumn'])
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of the ratings
plt.figure(figsize=(10, 6))
df['Ratings'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```
